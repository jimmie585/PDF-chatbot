```python
from langchain.document_loaders import PyPDFLoader

# Use only the file name (no Windows path)
loader = PyPDFLoader("james NLP.pdf")
docs = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Splitting the pdf into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(docs)

# Optional: View one chunk
print(chunks[0].page_content)
```

```python
# Embedding and vector storage FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Create embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create vector store
vectorstore = FAISS.from_documents(chunks, embeddings)
retriever = vectorstore.as_retriever()
```

```python
# loading the LLM and creating the RAG pipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load FLAN-T5 model
model_name = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Create pipeline
flan_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
```

```python
#Defining the RAG function
def query_rag(question):
    relevant_docs = retriever.get_relevant_documents(question)
    context = "\n".join([doc.page_content for doc in relevant_docs])
    prompt = f"Answer the question using only the context:\n\nContext:\n{context}\n\nQuestion: {question}\n\nAnswer:"
    response = flan_pipeline(
        prompt,
        max_new_tokens=200,
        temperature=0.9,
        top_k=50,
        top_p=0.9,
        do_sample=True
    )
    return response[0]['generated_text']
```

```python
# example
question = "Summarize the key points of this document in a paragraph of 200 words."
print(query_rag(question))
```
